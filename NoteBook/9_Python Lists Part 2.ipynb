```python
l = [1,25,12,4,6,8,74,3]

n = int(input("Enter a number to be searched: "))
i = 0
#           8
while i < len(l):
    if n == l[i]:
        print("Search Success")
    else:
        print("Fail")
    i += 1
```

```python
l = [1,25,12,4,6,8,74,3]

n = int(input("Enter a number to be searched: "))
i = 0

f = 0              # flag bit variable

#           8
while i < len(l):
    if n == l[i]:
        f = 1
        break
    i += 1

if f == 1:
    print("Search Success")
else:
    print("Fail")
```
