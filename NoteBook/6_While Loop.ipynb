### Reverse of a number

```python
n = int(input("Enter a number: "))
s = 0

while n > 0:               # n = 153, 15, 1
    r = n % 10             # r = 3, 5, 1
    s = s * 10 + r         # s = 0, 3, 35, 351
    n = n // 10            # n = 15, 1, 0

print("Reverse =", s)
```

**WAP to enter a number from user and check if it is palindrome or not**

153  -  351

121  -  121

```python
n = int(input("Enter a number: "))
temp = n
s = 0

while n > 0:               # n = 153, 15, 1
    r = n % 10             # r = 3, 5, 1
    s = s * 10 + r         # s = 0, 3, 35, 351
    n = n // 10            # n = 15, 1, 0

if s == temp:
    print("Palindrome")
else:
    print("Non Palindrome")
```

```python
n = int(input("Enter a number: "))

s = 0

while n > 0:               # n = 153
    r = n % 10             # r = 3, 5, 1
    s = s*10 + r           # s = 0, 27, 152, 153
    n = n // 10            # n = 15, 1

while s > 0:
    r = s % 10
    if r == 1:
        print("One")
    elif r == 2:
        print("Two")
    elif r == 3:
        print("Three")
    elif r == 4:
        print("Four")
    elif r == 5:
        print("Five")
    elif r == 6:
        print("Six")
    elif r == 7:
        print("Seven")
    elif r == 8:
        print("Eight")
    elif r == 9:
        print("Nine")
    else:
        print("Zero")
    s = s // 10
```

**WAP for factorial of a number**

```python
5!  - 5*4*3*2*1 - 120
n   1*n*n-1*n-1
       f
1! - 1
0! - 1
```

```python
n = int(input("Enter a number"))

f = 1

while n > 1:
    f = f * n    # f = 1, 5, 20, 60, 120
    n -= 1       # n = 5, 4, 3, 2, 1

print("Factorial =", f)
```
